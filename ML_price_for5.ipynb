```python
import pandas as pd
import numpy as np
```

```python
megh_data = pd.read_csv('apple.csv')
#megh_data[['asd1','asd2']] = megh_data.laptop_weight.str.split(" ")
megh_data.columns
```

```python
megh_data.drop(columns={'laptop_size','Screen_resolution','CPU_manufacturer','CPU_series','cache_memory','hard_disk','internal_memory'},inplace=True)
```

```python
#megh_data = pd.get_dummies(megh_data,columns=['manufactor','laptop_size','laptop_weight','laptop_type','Screen_size','Screen_resolution','CPU_manufacturer','CPU_series','cache_memory','RAM','hard_disk','internal_memory'],drop_first=True)
megh_data = pd.get_dummies(megh_data,columns=['manufactor','laptop_weight','laptop_type','Screen_size','RAM'],drop_first=True)

megh_data
```

```python
megh_data.drop(columns={'name','ename','sprice','Screen_type','CPU_model','CPU_frequency','RAM_type','graphic_manufacturer','graphic_model','graphic_memory','Laptop_id'},inplace=True)
megh_data
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X = megh_data.drop(['aprice'], axis=1)
y = megh_data['aprice']


x_train,x_test , y_train,y_test = train_test_split(X,y, test_size=.2, random_state=313)
```

```python
stan_scaler = StandardScaler()
x_train = stan_scaler.fit_transform(x_train)
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import r2_score
y_pred = model.predict(stan_scaler.transform(x_test))

r2 = r2_score(y_test, y_pred)
r2 
```
